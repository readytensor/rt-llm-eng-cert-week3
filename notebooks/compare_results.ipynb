```python
import json

baseline_predictions_path = './evaluate-baseline/predictions.jsonl'
finetuned_predictions_path = './evaluate_qlora/predictions.jsonl'



with open(baseline_predictions_path, 'r') as f:
    baseline_predictions = [json.loads(line) for line in f]

with open(finetuned_predictions_path, 'r') as f:
    finetuned_predictions = [json.loads(line) for line in f]



```

```python
def print_sample(index):
    dialogue = baseline_predictions[index]["dialogue"]
    reference = baseline_predictions[index]["reference"]
    print(f"Dialogue: {dialogue}")
    print('='*80)
    print(f"Reference: {reference}")
    print('\n'+'='*80)
    print(f"Baseline: {baseline_predictions[index]['prediction']}")
    print('\n'+'='*80)
    print(f"Finetuned: {finetuned_predictions[index]['prediction']}")
    print('\n'+'='*80)

    
```

```python
print_sample(0)
```

```python
print_sample(10)
```
